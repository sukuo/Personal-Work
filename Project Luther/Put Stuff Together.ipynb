# Import CSV Files

```python
grahamDF = pd.read_csv('script_df_for_sunny.csv')
sethDF = pd.read_csv('movie_mojo.csv')
ratingsDF = pd.read_csv('ratingsDF.csv')
```

# Cleaning and reformatting

```python
def rem_char(word):
    return (''.join(e for e in word if e.isalnum())).lower()

def rem_specs(x):
    x = re.sub('\D', '', x)
    return x

def other_genre(x):
    if x in lc_genre:
        return "other"
    else:
        return x
    
def action_genre(t):
    if t in action_genres:
        return 'action'
    else:
        return t
    
def eval_dict(x):
    return dict(eval(x))
```

```python
grahamDF['title'] = grahamDF['title'].apply(rem_char)
ratingsDF['title'] = ratingsDF['title'].apply(rem_char)
sethDF['title'] = sethDF['title'].apply(rem_char)
```

```python
merged_df = pd.merge(grahamDF, sethDF, on='title')
merged_dff = pd.merge(merged_df, ratingsDF, on='title')
merged_dff['lifetime_gross'] = merged_dff['lifetime_gross'].apply(rem_specs)
merged_dff['opening'] = merged_dff['opening'].apply(rem_specs)
merged_dff['theaters'] = merged_dff['theaters'].apply(rem_specs)
merged_dff['theaters.1'] = merged_dff['theaters.1'].apply(rem_specs)
#merged_dff['lifetime_gross'] = merged_dff['lifetime_gross'].apply(rem_doll)
#print(len(merged_df))

print(len(merged_dff))
#merged_dff.head(2)
#merged_dff['genre'].head(10)
```

# Clean up Genres

```python
least_common = dict(Counter(merged_dff['genre']))
lc_genre = []
for key, value in least_common.items():
    if value < 6:
        lc_genre.append((key))
```

```python
merged_dff['genre'] = merged_dff['genre'].apply(other_genre)
merged_dff['genre'] = merged_dff['genre'].apply(action_genre)
```

```python
final_df = pd.get_dummies(merged_dff, columns = ['genre'])
final_df['movie_script'] = final_df['movie_script'].apply(eval_dict)
#len(final_df)
final_df.head(2)
```
